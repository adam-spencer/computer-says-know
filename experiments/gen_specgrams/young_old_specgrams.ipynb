```python
def next_pow2(n: int) -> int:
    pow2 = 2
    while pow2 < n:
        pow2 *= 2
    return pow2
```

```python
plt.figure(figsize=(10,10), layout='tight')
plt.subplot(2,1,1)
plt.specgram(sig1, Fs=fs, NFFT=512, noverlap=32) #, pad_to=next_pow2(sig1.size))
plt.title('24 Year Old')
plt.subplot(2,1,2)
plt.specgram(sig2, Fs=fs, NFFT=512, noverlap=256) #, pad_to=next_pow2(sig2.size))
plt.title('83 Year Old')

None
```
